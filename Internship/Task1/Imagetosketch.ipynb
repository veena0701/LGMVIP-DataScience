```python
import matplotlib.pyplot as plt 
import cv2
```

```python
image = cv2.imread('dog.jpg')
type(image)
```

```python
# Resizing original image
im_resized = cv2.resize(image, (560, 560), interpolation=cv2.INTER_LINEAR)
```

```python
#displaying image in colour
plt.imshow(cv2.cvtColor(im_resized, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#converting colour image to grey scale and displaying
image_grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(image_grey, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#inverting the gray scale image and displaying
image_invert = cv2.bitwise_not(image_grey)
plt.imshow(cv2.cvtColor(image_invert, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#to blur the inverted image using GaussianBlur
blur = cv2.GaussianBlur(image_invert, (21, 21), 0)

plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#Inverting the blurred image again 
blur_invert = 255-blur

plt.imshow(cv2.cvtColor(blur_invert, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#The image to sketch version of the actual image
image_to_sketch = cv2.divide(image_grey, blur_invert, scale=256.0)
plt.imshow(cv2.cvtColor(image_to_sketch, cv2.COLOR_BGR2RGB))
plt.show()
```
